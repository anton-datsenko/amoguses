```python
import pandas as pd
df = pd.read_excel(r'proshares_analysis_data.xlsx', sheet_name='hedge_fund_series', index_col=0)
df
```

```python
sum_stat = pd.DataFrame(index=df.columns)
sum_stat["Mean"], sum_stat["Volatility"], sum_stat["Sharpe ratio"] = df.mean(), df.std(), df.mean()/df.std()
sum_stat
```

```python
cum_ret = (df + 1).cumprod()
cum_ret
```

```python
rolling_max = cum_ret.cummax()
rolling_max
```

```python
drawdown = (cum_ret-rolling_max)/rolling_max
drawdown
```

```python
drawdown.min()
```

```python
drawdown_dates = pd.DataFrame()
for i in drawdown.columns:
    drawdown_dates.loc[i, "Start Date"] = drawdown[i][drawdown[i] == 0].loc[:drawdown[i].idxmin()].index[-1]
drawdown_dates['End Date'] = drawdown.idxmin()
drawdown_dates
```

```python
for i in drawdown.columns:
    if drawdown[i][drawdown[i] == 0].loc[drawdown[i].idxmin():].size != 0:
        drawdown_dates.loc[i, "Recovery Date"] = drawdown[i][drawdown[i] == 0].loc[drawdown[i].idxmin():].index[0]

drawdown_dates
```

```python
drawdown_dates["Recovery Duration"] = drawdown_dates["Recovery Date"] - drawdown_dates["End Date"]
drawdown_dates
```

```python
forecast_expanded_mean = df.expanding().mean()
forecast_expanded_mean
```

```python
b_forecast = forecast_expanded_mean.shift(periods=1, fill_value=0)
b_forecast
```

```python
summ = 0
l = []
for col in b_forecast.columns:
  for ind in b_forecast.index:
    summ  += (forecast_expanded_mean.loc[ind, col] - b_forecast.loc[ind, col])**2
    l.append(summ/b_forecast.shape[0])
  print("MSE for", col, "column is", round(summ/b_forecast.shape[0], 8))
print("Total MSE for the whole dataset is", round(sum(l), 8))
```
